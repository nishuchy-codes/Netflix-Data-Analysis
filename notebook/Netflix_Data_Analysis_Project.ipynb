 # Netflix Content Strategy Analysis  

### Introduction:
This project is an **Exploratory Data Analysis (EDA)** of the Netflix Dataset. The goal is to analyze the distribution of content types (Movies vs TV Shows) and identify top contributing countries. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('netflix_titles.csv')
df.head()
```

### 1.Data Quality Check

```python
df.isnull().sum()
```

**Note:** We observed missing values in 'Director' and 'Cast' columns.This is common in such datasets.We will proceed with 'Type' and "country' for our main analysis.

### 2.Global Distribution of Movies vs TV Shows
This chart compares the total number of Movies and TV Shows available on Netflix globally. It helps us see which format is more dominant on the platform.

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='type', palette='viridis')
plt.title('Global Distribution: Movies vs TV Shows')
plt.savefig('5_global_distribution.png', dpi=300, bbox_inches ='tight')
plt.show()
```

### 3. Top 5 Content Producing Countries 
This visualization identifies the leading countries contributing to  Netflix's library. It highlights the major regional hubs for content production.

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df, y='country', order=df['country'].value_counts().index[:5], palette='magma', hue='country', legend=False)             
plt.title('Top 5 countries on netflix')
plt.savefig('top_5_countries.png',dpi=300, bbox_inches='tight')
plt.show()
```

### 4.Netflix Content Trend in India
By filtering the dataset for **India**,we analyze how the Indian market differs from global trends in terms of content preeference.  

```python
india_data = df[df['country'] == 'India']
plt.figure(figsize=(9,5))
sns.countplot(data=india_data, x='type', palette='Set2')
plt.title('Movies vs Tv shows in India')
plt.savefig('india_content_trend.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Final Conclusion & Insights:
* **Content Dominant:** Globally,Netflix is a movie-heavy platform, with Movies outnumbering TV Shows significantly.
* **India's Standing :** India is a major hub for Netflix content, ranking second globally.
* **Indian Market Insight:** The Indian audience shows an overwhelming preference for Movies over TV Shows, even more than the global average.
